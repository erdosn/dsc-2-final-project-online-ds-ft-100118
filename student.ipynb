```python
import sqlite3
import sqlalchemy
import pandas as pd
```

```python
sql = sqlite3.connect("./Northwind_small.sqlite")
```

```python
table_names = [tup[0] for tup in sql.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()]
table_names
```

```python
dfs = {}
for table_name in table_names:
    df = pd.read_sql_query("select * from [{}]".format(table_name), sql)
    dfs[table_name]=df
```

```python
for table_name, table in dfs.items():
    print(table_name.upper())
    print(table.info())
    print("\n\n\n")
```

```python
def triangle(n):
    if n == 0: 
        return 0
    if n == 1: 
        return 1
    else:
        return n + triangle(n-1)
```
